1. Introducci√≥n
1.1 Planteamiento del problema

Las empresas de comercio electr√≥nico necesitan comprender y predecir el comportamiento de sus usuarios y la din√°mica de sus pedidos para optimizar inventario, log√≠stica, campa√±as y pol√≠ticas de precios. En particular, disponer de un dataset que combine informaci√≥n de pedidos, items, pagos, clientes y productos permite responder preguntas cr√≠ticas del negocio: ¬øqu√© factores influyen en la conversi√≥n?, ¬øqu√© productos generan mayor valor?, ¬øhay patrones geogr√°ficos de demanda?, y ¬øc√≥mo afectan las formas de pago al valor promedio de pedido?

El objetivo principal es transformar esas tablas en una √∫nica vista tabular lista para an√°lisis exploratorio y modelado predictivo, y usarla para construir modelos que ayuden a estimar m√©tricas de inter√©s (por ejemplo: probabilidad de compra repetida, predicci√≥n de valor de pedido, clasificaci√≥n de productos por probabilidad de devoluci√≥n, etc.).

https://github.com/DavidGN2016GFS/ecommerce-order-analysis/releases/download/v1.0/ecommerce_merged_full.csv

```python
# =============================================================
# üß© Secci√≥n 4 ‚Äî Limpieza y Preprocesamiento Mejorado
# =============================================================

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.feature_selection import SelectKBest, chi2

print("üöÄ Iniciando preprocesamiento inteligente...")

# ------------------------------------------------------------
# 4.1 Carga del dataset
# ------------------------------------------------------------
url = "https://github.com/DavidGN2016GFS/ecommerce-order-analysis/releases/download/v1.0/ecommerce_merged_full.csv"
df = pd.read_csv(url, low_memory=False)
print(f"üì¶ Dataset original: {df.shape[0]} filas, {df.shape[1]} columnas\n")

# ------------------------------------------------------------
# 4.2 Limpieza b√°sica
# ------------------------------------------------------------
print("üßπ Eliminando duplicados y manejando nulos...")

df = df.drop_duplicates()
print(f"üî∏ Filas despu√©s de eliminar duplicados: {df.shape[0]}")

cols_criticas = [c for c in ['price', 'payment_value', 'product_category_name'] if c in df.columns]
if cols_criticas:
    df = df.dropna(subset=cols_criticas)
    print(f"üî∏ Filas despu√©s de eliminar nulos cr√≠ticos: {df.shape[0]}")
else:
    print("‚ö†Ô∏è No se encontraron columnas cr√≠ticas esperadas.")

# Rellenar nulos con mediana o moda
for col in df.select_dtypes(include=[np.number]):
    df[col].fillna(df[col].median(), inplace=True)
for col in df.select_dtypes(include=['object']):
    if df[col].isnull().any():
        df[col].fillna(df[col].mode()[0], inplace=True)

print("‚úÖ Limpieza b√°sica completada.\n")

# ------------------------------------------------------------
# 4.3 Eliminaci√≥n de columnas irrelevantes
# ------------------------------------------------------------
cols_to_drop = [
    'order_id', 'customer_id', 'seller_id',
    'order_purchase_timestamp', 'order_approved_at',
    'customer_zip_code_prefix'
]
df.drop(columns=cols_to_drop, inplace=True, errors='ignore')
print(f"üóëÔ∏è Columnas eliminadas: {cols_to_drop}\n")

# ------------------------------------------------------------
# 4.4 Codificaci√≥n de variables categ√≥ricas
# ------------------------------------------------------------
print("üî† Codificando variables categ√≥ricas...")

cat_cols = df.select_dtypes(include=['object']).columns.tolist()
print(f"üî∏ Columnas categ√≥ricas detectadas: {cat_cols}")

# Guardar backup de columnas antes del one-hot
before_encoding = set(df.columns)

# One-Hot Encoding para baja cardinalidad
low_cardinality = [c for c in cat_cols if df[c].nunique() <= 10]
if low_cardinality:
    df = pd.get_dummies(df, columns=low_cardinality, drop_first=True)
    print(f"‚úÖ One-Hot Encoding aplicado a: {low_cardinality}")

# LabelEncoder para alta cardinalidad
high_cardinality = [c for c in cat_cols if c not in low_cardinality]
for col in high_cardinality:
    if col in df.columns:
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
        print(f"üîπ Codificada (LabelEncoder): {col}")

print("‚úÖ Codificaci√≥n completada.\n")

# ------------------------------------------------------------
# 4.5 Escalado de variables num√©ricas
# ------------------------------------------------------------
print("üìè Escalando variables num√©ricas...")
num_cols = df.select_dtypes(include=[np.number]).columns
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
print("‚úÖ Escalado (StandardScaler) completado.\n")

# ------------------------------------------------------------
# 4.6 Selecci√≥n del target autom√°ticamente
# ------------------------------------------------------------
print("üéØ Identificando variable objetivo (target)...")

# Lista de posibles targets
possible_targets = ['payment_type', 'order_status', 'review_score', 'payment_value']

target = None
for t in possible_targets:
    if t in before_encoding:  # si exist√≠a antes de los dummies
        target = t
        print(f"‚úîÔ∏è Target detectado: {target}")
        break

if target is None:
    print("‚ö†Ô∏è No se encontr√≥ target expl√≠cito. El proceso seguir√° sin selecci√≥n de variables.")
else:
    # ------------------------------------------------------------
    # 4.7 Selecci√≥n de caracter√≠sticas (Chi-cuadrado)
    # ------------------------------------------------------------
    print("üîç Seleccionando caracter√≠sticas m√°s relevantes con Chi¬≤...")

    # Reconstruir variable target (si fue convertida a dummies)
    if target not in df.columns:
        possible_dummies = [c for c in df.columns if target in c]
        if possible_dummies:
            print(f"‚úÖ Usando dummies relacionadas con {target}: {possible_dummies}")
            y = df[possible_dummies].idxmax(axis=1)
            y = LabelEncoder().fit_transform(y)
        else:
            print(f"‚ö†Ô∏è No se encontraron dummies de {target}, se omite selecci√≥n.")
            y = None
    else:
        y = df[target]
        if y.dtype == 'object':
            y = LabelEncoder().fit_transform(y)

    if y is not None:
        X = df.drop(columns=[c for c in df.columns if target in c], errors='ignore')
        X_pos = X.copy()
        X_pos[X_pos < 0] = 0  # chi¬≤ solo acepta positivos

        selector = SelectKBest(chi2, k=min(10, X_pos.shape[1]))
        selector.fit(X_pos, y)
        selected_features = X.columns[selector.get_support()]
        print("‚úÖ Variables m√°s relevantes seg√∫n Chi¬≤:")
        print(selected_features.tolist())

# ------------------------------------------------------------
# 4.8 Guardar dataset procesado
# ------------------------------------------------------------
output_path = "ecommerce_processed.csv"
df.to_csv(output_path, index=False)
print(f"\nüíæ Dataset procesado guardado como '{output_path}'")
print("üìå Pr√≥ximo paso: ejecutar la Secci√≥n 5 ‚Äî Modelado (SVM, Random Forest, ANN).")
```
